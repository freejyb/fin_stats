```python
from fredapi import Fred
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
import json
# 한글폰트 출력 가능 코드
from matplotlib import rc
rc('font', family='AppleGothic') # Mac Os
#rc('font', family='NanumGothic') # Windows Os
plt.rcParams['axes.unicode_minus'] = False

# FRED api 키 입력 (https://fred.stlouisfed.org)
fred_key='개인 key 입력'

# ECOS API 키 입력 (https://ecos.bok.or.kr/jsp/openapi/OpenApiController.jsp)
ecos_key = '개인 key 입력'
```

# FRED API

```python
# FRED api 키 입력
fred = Fred(api_key = fred_key)

# S&P 500 불러오기
sp500 = fred.get_series('SP500')
sp500 = pd.DataFrame(sp500, columns=['SP500'])
sp500
```

```python
# 장단기 금리 차이 (10년, 2년물)
t10y2y = fred.get_series('T10Y2Y', observation_start='4/30/2012', observation_end='4/27/2022')
t10y2y = pd.DataFrame(t10y2y, columns=['t10y2y'])
t10y2y
```
