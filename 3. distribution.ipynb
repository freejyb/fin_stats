## A. 이항분포

```python
# 필요한 라이브러리 호출 (팩토리얼 계산을 위해 math의 factorial 매서드를 호출함)
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

# Probability density of the binomial distribution
# P만큼의 성공확률을 갖는 n 회 시행에 대해서 k회가 성공할 확률을 도출함
def bin_dist(k, n, p):
    nck = factorial(n) / (factorial(k) * factorial(n - k))
    proba_df = nck * pow(p, k) * pow(1-p, n-k)
    return proba_df


# 이항분포 형성
# 변수 설정
n = 15
p = 0.3

# x는 0 ~ 15의 값을 가지는 리스트
x = np.arange(16)
print(x)

# 이항분포 확률을 나타내는 리스트로 성공이 0.3의 확률일 때, 15번 시행한 경우, k번 성공할 확률을 리스트로 나타냄
proba_df1 = np.array([bin_dist(k, n, p) for k in range(16)])
print(proba_df1)

# 표를 그리는 코드
plt.ylim(0, 0.3)
plt.bar(x, proba_df1, color='lightcoral', label = f'independent experiments : {n}, Success proba : {p * 100}%')
plt.ylabel('확률')
plt.xlabel('성공횟수')
plt.legend()
plt.show()
```

## B. 베르누이 확률분포

- 동전던지기

```python
# 라이브러리 호출
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# 성공확률 설정
p = 0.5

rv = bernoulli(p)

XX = [0, 1]
plt.bar(XX, rv.pmf(XX), align="center")
plt.ylim(0,1)
plt.xticks([0, 1], ["X=0", "X=1"])
plt.ylabel("Proba")
plt.title("Bernoulli df")
plt.show()
```

```python
# 함수화
def bernoulli_df(p):
    # 라이브러리 호출
    import matplotlib.pyplot as plt
    from scipy.stats import bernoulli

    proba = bernoulli(p)

    x = [0, 1]
    plt.bar(x, proba.pmf(x), align="center")
    plt.ylim(0,1)
    plt.xticks([0, 1], ["X=0", "X=1"])
    plt.ylabel("Proba")
    plt.title("Bernoulli df")
    plt.show()
    
bernoulli_df(0.1)
```

```python
import seaborn as sns
from scipy.stats import bernoulli
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Bernoulli 분포 매개변수
p = 0.3
n = 1000

# Bernoulli 분포 샘플 생성
rv = bernoulli(p)
X = rv.rvs(n, random_state=2022)
y = np.bincount(X, minlength=2) / len(X)
XX = [0, 1]

# 시뮬레이션 결과 히스토그램
sns.countplot(x=X)
plt.xticks(
    [0, 1],
    [f"X=0\n{int(y[0] * n)} times", f"X=1\n{int(y[1] * n)} times"]
)
plt.title("Simulation of Bernoulli Distribution")
plt.show()

# 이론적 분포와 시뮬레이션 분포 비교 데이터 생성
df = pd.DataFrame({
    "value": XX,
    "theoretic": rv.pmf(XX),
    "simulation": y
})
df = df.melt(id_vars="value", var_name="type", value_name="ratio")

# 이론적 분포와 시뮬레이션 분포 비교 시각화
sns.barplot(x="value", y="ratio", hue="type", data=df)
plt.title("Simulation vs Theoretic Distribution")
plt.show()
```

```python
def sim_bernoulli_df(p, n):
    import seaborn as sns
    from scipy.stats import bernoulli
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt

    # Bernoulli 분포 생성 및 샘플링
    rv = bernoulli(p)
    X = rv.rvs(n, random_state=2022)
    y = np.bincount(X, minlength=2) / len(X)

    # 시각화 - Bernoulli 분포 샘플링 결과
    sns.countplot(x=X)
    plt.xticks(
        [0, 1],
        [f"X=0\n{int(y[0] * n)} times", f"X=1\n{int(y[1] * n)} times"]
    )
    plt.title(f"Simulation of Bernoulli Distribution ({n} times)")
    plt.show()

    # 이론적 분포와 시뮬레이션 분포 비교
    XX = [0, 1]
    df = pd.DataFrame({"value": XX, "theoretic": rv.pmf(XX), "simulation": y})
    df = df.melt(id_vars="value", var_name="type", value_name="ratio")

    # 시각화 - 이론적 분포와 시뮬레이션 비교
    sns.barplot(x="value", y="ratio", hue="type", data=df)
    plt.title(f"Simulation vs Theoretic ({n} times)")
    plt.show()

# 함수 호출
sim_bernoulli_df(0.3, 10000)
```

## A. 정규분포, 표준정규분포

```python
# 정규분포 그리는 코드 (예시)

# 라이브러리 호출
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns

# 정규분포를 이루는 데이터 조건 형성
# 데이터 개수 500개, 분산, 평균 설정
N1 = 500
sigma_1 = 10
mu_1 = 10

# 랜덤데이터 형성 (정규분포)
np.random.seed(0)
x1 = sp.stats.norm(mu_1, sigma_1).rvs(N1)

# 시각화
ax = sns.distplot(x1, kde=False, fit=sp.stats.norm, label=f"normal df (Variance :{np.round(x1.std(),3)})")
plt.legend(bbox_to_anchor=(1.0, .2))
plt.show()
```

```python
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc
from matplotlib import font_manager

import matplotlib.pyplot as plt
from matplotlib import rc

# 맑은 고딕 폰트 설정
rc('font', family='Malgun Gothic')

# 정규분포 그리는 함수
def normal_df(N, sigma, mu):
    # 랜덤 데이터 형성 (정규분포)
    np.random.seed(0)
    x = sp.norm(mu, sigma).rvs(N)

    # 시각화 (mu=0으로 입력하는 경우 x-1을 넣어서 에러 해결)
    sns.histplot(x - 1, kde=True, stat="density", label=f"Normal DF (Variance: {x.std() :.3f})")
    plt.legend(bbox_to_anchor=(1.0, 0.2))
    plt.ylabel('PDF')
    plt.xlabel('X')
    if (mu == 1) & (sigma == 1):
        plt.title('SND 표준정규분포')
    else:
        plt.title('N-D')
    plt.show()

# 정규분포 형성 (평균 = 10, 표준편차 = 10)
normal_df(1000000, 10, 10)

# 표준정규분포 형성 (평균 = 1, 표준편차 = 1)
normal_df(1000000, 1, 1)
```

## B. T 분포

```python
# 라이브러리 호출
import math
import numpy as np
import matplotlib.pyplot as plt

# t 정의 (-5, 5 까지 101개의 값으로 분류 (0.1 간격))
t = np.linspace(-5, 5, 101)

# 자유도 설정
DF = 10
y = (math.gamma((DF+1) / 2) / (math.gamma(DF / 2) * math.sqrt(math.pi * DF))) * (1 + t ** 2 / DF) ** (-(DF + 1) / 2)

plt.plot(t, y)                       
plt.xlabel("t")                    
plt.ylabel("pdf")                      
plt.title("t-df")
plt.legend([f"t-df (DF = {DF})"])
plt.show()
```

```python
def t_df(DF):
    # 라이브러리 호출
    import math
    import numpy as np
    import matplotlib.pyplot as plt

    t = np.linspace(-5, 5, 101)
    y = (math.gamma((DF+1) / 2) / (math.gamma(DF / 2) * math.sqrt(math.pi * DF))) * (1 + t ** 2 / DF) ** (-(DF + 1) / 2)

    plt.plot(t, y)                       
    plt.xlabel("t")                    
    plt.ylabel("pdf")                      
    plt.title("t-df")
    plt.legend([f"t-df (Degree of Freedom = {DF})"], bbox_to_anchor=(1, .2))
    plt.show()
    
t_df(5)
```

```python
# scipy의 stats 호출로 t 분포 그리기
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

DF = 10
t = np.linspace(-5, 5, 101)
y1 = stats.t(DF).pdf(t)
       
plt.plot(t, y1)
plt.xlabel("t")
plt.ylabel("pdf")
plt.title("t-df with scipy")
plt.legend([f"t-df (DF = {DF})"])
plt.show()
```

```python
def t_df_stats(DF):
    import scipy.stats as stats
    import numpy as np
    import matplotlib.pyplot as plt

    t = np.linspace(-5, 5, 101)
    y1 = stats.t(DF).pdf(t)
        
    plt.plot(t, y1)
    plt.xlabel("t")
    plt.ylabel("pdf")
    plt.title("t-df with scipy")
    plt.legend([f"t-df (Degree of Freedom = {DF})"])
    plt.show()
    
t_df_stats(5)
```

## C. F 분포

```python
import scipy.special as sp
import math
import numpy as np

# X축 값 설정 (0,5 사이에 일정한 거리로 1000개 데이터 입력)
x = np.linspace(0, 5, 1000)
# 각각의 독립적인 확률변수 데이터가 자유도 4, 10인 카이제곱 분포를 따르는 경우를 가정
nu1 = 4
nu2 = 10

# F분포를 따르는 확률변수를 생성
y = []
for f in x:
    y.append(( np.sqrt( ((nu1 * f) ** nu1 * nu2 ** nu2) / ((nu1 * f + nu2) ** (nu1 + nu2)) )) / 
            (f * sp.beta(nu1 / 2, nu2 / 2)))
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, label=f'F({nu1}, {nu2})')                       
plt.xlabel('F')
plt.ylabel("y")
plt.title('F 분포 without scipy')
plt.legend()
plt.show()
```

```python
# scipy를 이용한 F분포 형성
import scipy.stats as stats

# 카이제곱분포를 따르는 확률변수의 자유도가 nu1, nu2일 때 F분포의 확률밀도함수는 y1으로 정의된다.
y1 = stats.f(nu1, nu2).pdf(x)        

plt.plot(x, y1, color="g", label='F(4, 50)')             
plt.xlabel('F')
plt.ylabel("y")
plt.title('scipy를 이용한 F 분포')
plt.legend()
plt.show()
```

## D. 카이제곱 분포

```python
# 라이브러리 호출
import numpy as np
import scipy.stats as stats  # 올바른 scipy.stats 임포트
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc
from matplotlib import font_manager


# 자유도 설정
DF = 4

# 데이터 형성
X = np.linspace(0, 15, 151)

# 시각화
plt.plot(X, stats.chi2(DF).pdf(X), label=f"Chi2 (DF = {DF})")  # stats.chi2로 수정
plt.legend()
plt.title('Chi-Square df')
plt.show()


```

```python
def chi2(DF):
    import numpy as np
    from scipy.stats import chi2
    import matplotlib.pyplot as plt

    X = np.linspace(0, 15, 151)

    plt.plot(X, chi2(DF).pdf(X), label=f"Chi2 (Degree of Freedom = {DF})")
    plt.legend()
    plt.title('Chi-Square df')
    plt.show()
    
chi2(4)
```
