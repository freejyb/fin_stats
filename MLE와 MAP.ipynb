```python
import numpy as np
import matplotlib.pyplot as plt

# Observed data: 7 heads out of 10 coin flips
n = 10
k = 7

# Define a range of p values from 0 to 1
p_values = np.linspace(0, 1, 1000)

# Calculate the likelihood function for MLE
likelihoods = [(np.math.comb(n, k)) * (p**k) * ((1 - p)**(n - k)) for p in p_values]

# Define a prior for MAP (e.g., Beta distribution with alpha=2, beta=2)
alpha_prior = 2
beta_prior = 2
priors = [(p**(alpha_prior - 1)) * ((1 - p)**(beta_prior - 1)) for p in p_values]

# Calculate the posterior for MAP (unnormalized)
posteriors = [likelihood * prior for likelihood, prior in zip(likelihoods, priors)]

# Plot the results
plt.figure(figsize=(12, 6))

# Likelihood for MLE
plt.plot(p_values, likelihoods, label='Likelihood (MLE)', color='blue')

# Posterior for MAP
plt.plot(p_values, posteriors, label='Posterior (MAP)', color='red')

plt.xlabel('p (Probability of Success)')
plt.ylabel('Value')
plt.title('MLE vs. MAP for Binomial Distribution (n=10, k=7)')
plt.axvline(x=0.7, color='blue', linestyle='--', label='MLE Estimate (p=0.7)')
plt.axvline(x=0.666, color='red', linestyle='--', label='MAP Estimate (p~0.666)')
plt.legend()
plt.grid(True)
plt.show()
```
