```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 기본 정리

1. plot은 실제 차트를 생성하는 것이 아닌 내부적으로 캠퍼스를 형성하는 것임 (따라서 plt.show()를 입력해야 함)
2. Title, Label
   1. plt.title('Title') : 차트의 제목 설정
   2. plt.xlabel('X name') : 차트의 x 축 이름 설정
   3. plt.ylabel('Y name') : 차트의 y 축 이름 설정
3. Grid
   1. plt.grid(axis = {'y','x','both'})
4. range
   1. plt.xlim() : x 축 범위 설정
   2. plt.ylim() : y 축 범위 설정
5. style
   1. plt.plot(color = {#RGB 설정}, markers = {점 모양 설정}, line)
6. legend (범례)
   1. plt.plot(x, y, label = '범례 이름 설정')
   2. plt.legend(loc = 'best') : 제일 최적의 위치로 범례를 출력

```python
# x, y 지정
# numpy.linspace 함수는 지정한 구간을 균일한 간격으로 나누는 숫자들을 반환
# (start, num. stop)  ==>  start : 배열 시작값, stop : 배열의 끝값, num : 몇개의 구간
x = np.linspace(1,10,20)
y = x ** 2

# Color (색깔) : 약자 또는 컬러이름 또는 ‘#RGB’로 지정가능
# b : blue
# g : green
# r : red
# y : yellow
# k : black
# Markers (점 모양): 문자 형태로 지정가능
# . : point marker
# o : circle marker
# ^ : triangle_up marker
# * : star marker
# x : x marker
# D : diamond marker
# Line style (선 스타일) : 문자 형태로 지정가능
# - : solid line
# -- : dashed line
# -. : dash-dot line
# : : dotted line

plt.plot(x,y, color ='g', marker = '*', linestyle =':', label='x')
plt.title('y = $x^2$')
plt.xlabel('x')
plt.ylabel('$x^2$')
plt.legend(loc='best')
plt.grid(axis='both')
plt.xlim(0,20)
plt.ylim(0,200)
plt.show()
```

```python
# 예시 코드
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
```

# 같은 표에 두개의 그래프 출력

```python
x = np.arange(1,10,0.1)
y = x*0.2
y2 = np.sin(x)

# 같은 표 내에 두개의 그래프를 출력하는 경우
plt.plot(x,y,'b',label='y = 0.2x') 
plt.plot(x,y2,'g',label='y = sin(x)') 
plt.xlabel('x') 
plt.ylabel('y')
plt.title('sample plot')
plt.legend(loc='upper right')
plt.show()
```

# 서브플롯 활용하기

- 여러개의 그래프를 그리는 경우 사용한다. (대신 그래프는 한 플롯에 출력하는 것이 아닌 따로 출력)
1. plt.subplot(nrow,ncol,pos)
   1. nrow : 몇 개의 행을 출력하는지
   2. ncol : 몇 개의 열을 출력하는지
   3. pos : 몇 번째에 출력할 것인지 (상단부터 우측, 아래 순으로 출력)

```python
x = np.arange(1,10)
y1 = x*5
y2 = x*10

# 2,1의 칸 1번째에만 출력하는 경우
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.subplot(2,1,1)
plt.plot(x,y2, color='g')
plt.title('equal position')
plt.show()

# 서브플롯을 2,1 칸에 각각 출력하는 경우
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.ylabel('y1')
plt.title('subplot')
plt.subplot(2,1,2)
plt.plot(x,y2, color='g')
plt.ylabel('y2')
plt.xlabel('x')
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 50)

# plot 내에 x,y 정보를 더 넣으면 그만큼 또 출력된다.
plt.plot(x, np.sin(x), x, np.cos(x))
plt.show()
```

```python
# x,y 설정
x = np.linspace(0, 2*np.pi, 100) 
y = np.sin(x)

# 1번째 예시 그래프
plt.plot(x,y)
plt.title('1st')
plt.show()


# x, y 설정
x = np.linspace(0, 2*np.pi, 100) # [0,2PI] 사이 수 100개 생성. 0과 2PI 포함
y = np.sin(x)  # 여기서 x는 numpy의 array이고, np.sin(x)는 이 x 배열값 각각에 대한 값 계산

# 2번째 예시 그래프
fig, ax = plt.subplots(figsize=(10,6))  # subplots를 호출하면서, 그림 크기 지정도 가능함 
ax.set_xlim(0, 2*np.pi) # x축 범위
ax.set_ylim(-1.2, 1.2)  # y축 범위
ax.set_xticks([0,np.pi/2,np.pi,(3*np.pi)/2,2*np.pi]) # x축의 값을 특정 값으로 출력하도록 설정하는 부분 (xticks는 x축의 간격을 의미함)
ax.set_xticklabels(['0',r'$\dfrac{\pi}{2}$',r'$\pi$',r'$\dfrac{3\pi}{2}$',r'$2\pi$']) #dfrac에 유의. 폰트가 큰 분수
ax.grid(True) # 그래프의 눈금 격자 표시
ax.set_title('2nd')
plt.plot(x,y) # 디폴트 라인을 가지고 (x,y)에 대한 라인그래프를 그림
plt.show()
```

```python
# cumsum은 누적합 그래프
y1 = np.random.normal(size = 50).cumsum()
y2 = np.random.normal(size = 100).cumsum()
y3 = np.random.normal(size = 50).cumsum()
y4 = np.random.normal(size = 100).cumsum()

plt.subplot(2, 2, 1)
plt.plot(y1, label='y1')
plt.legend()

# same as plt.subplot(222)
plt.subplot(2, 2, 2)
plt.plot(y2, label='y2')
plt.legend()

# same as plt.subplot(2,2,3), plt.subplot(223)이라 써도 결과는 동일
plt.subplot(223)   
plt.plot(y3, label='y3')
plt.legend()

plt.subplot(224)
plt.plot(y4, label='y4')
plt.legend()
plt.show()
```
