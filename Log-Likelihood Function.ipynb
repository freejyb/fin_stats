```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the binomial distribution
n = 10  # Number of trials
k = 7   # Number of successes observed

# Define a range of p values from 0 to 1
p_values = np.linspace(0, 1, 1000)

# Calculate likelihood values for each p
likelihoods = [(np.math.comb(n, k)) * (p**k) * ((1 - p)**(n - k)) for p in p_values]

# Calculate log-likelihood values
log_likelihoods = [np.log(l) if l > 0 else -np.inf for l in likelihoods]

# Plot likelihood and log-likelihood
plt.figure(figsize=(14, 6))

# Plot likelihood
plt.subplot(1, 2, 1)
plt.plot(p_values, likelihoods, color='blue')
plt.xlabel('p')
plt.ylabel('Likelihood')
plt.title('Likelihood Function for Binomial Distribution')
plt.grid(True)

# Plot log-likelihood
plt.subplot(1, 2, 2)
plt.plot(p_values, log_likelihoods, color='red')
plt.xlabel('p')
plt.ylabel('Log-Likelihood')
plt.title('Log-Likelihood Function for Binomial Distribution')
plt.grid(True)

plt.tight_layout()
plt.show()
```
